```python
# Extract GameStop stock data
gamestop_stock = yf.Ticker('GME')
gamestop_data = gamestop_stock.history(period='1y')  # Fetch data for the past year

# Display the first few rows of the data
gamestop_data.head()
```

```python
import requests
from bs4 import BeautifulSoup

# URL for GameStop's financials on Yahoo Finance
url = 'https://finance.yahoo.com/quote/GME/financials?p=GME'

# Send GET request to fetch the page content
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the table containing financials
    tables = soup.find_all('table')
    if len(tables) > 1:  # Ensure there is a table for the financial data
        table = tables[1]  # The second table contains the revenue data

        # Loop through rows and extract relevant data
        gamestop_revenue = []
        for row in table.find_all('tr')[1:]:  # Skip the header row
            cols = row.find_all('td')
            if len(cols) > 1:  # Ensure there are at least two columns
                year = cols[0].text.strip()
                rev = cols[1].text.strip()
                gamestop_revenue.append([year, rev])

        # Print extracted revenue data
        for item in gamestop_revenue:
            print(item)
    else:
        print("Error: Could not find the revenue table on the webpage.")
else:
    print(f"Error: Unable to fetch the page. Status code: {response.status_code}")
```
